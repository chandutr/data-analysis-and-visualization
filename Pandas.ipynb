np.random.randint() is used to create random integers 40--> start value 100--> end value 6-->rows 4-->columns

```python
d=np.random.randint(40,100,(6,4))
d
```

pd.DataFrame() is used to create DataFrame using numpy module

```python
f=pd.DataFrame(d, columns=['rohan','harish','manu','chandu'], index=['First','second','third','fourth','fifth','sixth'])
f
```

creating dataframe from list of list 

```python
b=[['ab',2],['bc',3],['cd',4],['de',5]]
a=pd.DataFrame(b, columns=['name','marks'])
a
```

creating dataframe from dictionaries

```python
d = {'name': ['chandu', 'harish','rohan','manu'], 'marks': [98, 97,96,95]}
e = pd.DataFrame(d)
print(e)
```

```python
#creating dataframe from list of dictionary
k=[{'abc':2,'asd':5,'cvb':6},{'abc':7,'asd':8,'cvb':9}]
pd.DataFrame(k)
```
